#### **1. How many multiplications and additions do you need to perform a matrix multiplication between a (n, k) and (k, m) matrix? Explain.**

**1 Solution:** \\
Number of Multiplications = $ m \times n \times k $ <br>
Number of additions = $ m \times n \times (k-1) $ <br>

Given matrices are $ A_{(n,k)}, B_{(k,m)} $ the matrix multiplication will be : <br>
$ A \times B = \begin{bmatrix}
\sum_{i=1}^k a_{1i} b_{i1} &\sum_{i=1}^k a_{1i} b_{i2} & ... & \sum_{i=1}^k a_{1i} b_{im}\\
\sum_{i=1}^k a_{2i} b_{i1} &\sum_{i=1}^k a_{2i} b_{i2} & ... & \sum_{i=1}^k a_{2i} b_{im}\\
. & . & ... & .\\
. & . & ... & .\\
\sum_{i=1}^k a_{ni} b_{i1} &\sum_{i=1}^k a_{ni} b_{i2} & ... & \sum_{i=1}^k a_{ni} b_{im}\\
\end{bmatrix} $ <br>

So every element in the above matrix has $ k $ multiplications and $ k-1 $ additions.<br>
Total no of elements are $ n \times m $<br>
So,
Number of Multiplications = $ m \times n \times k $ <br>
Number of additions = $ m \times n \times (k-1) $

#### **10. Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as “a”, “b”, “c”, “d”, “e”. Find the column with the highest standard deviation. Find the row with the lowest mean.**

```python
import numpy as np
import pandas as pd

matrix = np.random.uniform(1,2, size = (20,5))
cols = ["a", "b", "c", "d", "e"]
df = pd.DataFrame(matrix, columns=cols)

std_devs = df.std()
print(f"Maximum standard deviation column = {std_devs.idxmax()}" )

mean_vals = df.mean(axis=1)
print(f"Minimum mean row = {mean_vals.idxmin()}")



```

#### **13. Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.**

```python
#13 solution:
import numpy as np
A = np.array([6, 2, 3, 4, 5])

def argmin(arr):
  min_ele = float('inf')
  min_index = 0
  for i in range(len(arr)):
    if arr[i] < min_ele:
      min_ele = arr[i]
      min_index = i
  return min_index

print(f"Minimum element index in A  = {argmin(A)}")
print(f"Minimum element index using np.argmin in A = {np.argmin(A)}")
```
