#### **1. How many multiplications and additions do you need to perform a matrix multiplication between a (n, k) and (k, m) matrix? Explain.**

**1 Solution:** \\
Number of Multiplications = $ m \times n \times k $ \\
Number of additions = $ m \times n \times (k-1) $

Given matrices are $ A_{(n,k)}, B_{(k,m)} $ the matrix multiplication will be : \\
$ A \times B = \begin{bmatrix}
\sum_{i=1}^k a_{1i} b_{i1} &\sum_{i=1}^k a_{1i} b_{i2} & ... & \sum_{i=1}^k a_{1i} b_{im}\\
\sum_{i=1}^k a_{2i} b_{i1} &\sum_{i=1}^k a_{2i} b_{i2} & ... & \sum_{i=1}^k a_{2i} b_{im}\\
. & . & ... & .\\
. & . & ... & .\\
\sum_{i=1}^k a_{ni} b_{i1} &\sum_{i=1}^k a_{ni} b_{i2} & ... & \sum_{i=1}^k a_{ni} b_{im}\\
\end{bmatrix} $ \\

So every element in the above matrix has $ k $ multiplications and $ k-1 $ additions. \\
Total no of elements are $ n \times m $ \\
So,
Number of Multiplications = $ m \times n \times k $ \\
Number of additions = $ m \times n \times (k-1) $

#### **2. Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?**

```python
import random
import numpy as np
import time

def create_random_matrix(rows, cols, min_val=0, max_val=9):
    return [[random.randint(min_val, max_val) for _ in range(cols)] for _ in range(rows)]

n, k, m = 3, 4, 5  # Dimensions for the matrices
A = create_random_matrix(n, k)
B = create_random_matrix(k, m)


# Performing matrix multiplication without numpy
start_time = time.time()
C = [[0 for _ in range(m)] for _ in range(n)]

for i in range(n):
  for j in range(m):
    C[i][j] = sum(A[i][l] * B[l][j] for l in range(k))
end_time = time.time()
print(f"Time taken without numpy: {end_time - start_time} seconds")

start_time = time.time()
C = np.dot(A, B)
end_time = time.time()
print(f"Time taken with numpy: {end_time - start_time} seconds")


```

$ \frac{\partial f}{\partial y} =  x^2 + 3y^2 \sin(x)$ \\

$ \nabla f =  2xy + y^3 \cos(x), x^2 + 3y^2 \sin(x)  $

#### **9 .Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.**

$ y = x $ \\
$ y = x^2 $ \\
$ y = \frac {x^3} {100} $ \\
$ y = sin(x) $ \\
$ y = \frac {sin(x)} {x} $ \\
$ y = log(x) $ \\
$ y = e^x $ \\






```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.5, 100.0, 0.5)

y1 = x
y2 = x**2
y3 = x**3 / 100
y4 = np.sin(x)
y5 = np.sin(x) / x
y6 = np.log(x)
y7 = np.exp(x)

plt.figure()
plt.plot(x,y1, label='y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$ y = x $')
plt.show()

plt.figure()
plt.plot(x,y2, label='y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$ y = x^2 $')
plt.show()

plt.figure()
plt.plot(x,y3, label='y = x^3/100')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$ y = x^3/100 $')
plt.show()

plt.figure()
plt.plot(x,y4, label='y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$ y = sin(x) $')
plt.show()

plt.figure()
plt.plot(x,y5, label='y = sin(x)/x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$ y = sin(x)/x $')
plt.show()

plt.figure()
plt.plot(x,y6, label='y = log(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$ y = log(x) $')
plt.show()

plt.figure()
plt.plot(x,y7, label='y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$ y = e^x $')
plt.show()
```

#### **13. Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.**

```python
#13 solution:
import numpy as np
A = np.array([6, 2, 3, 4, 5])

def argmin(arr):
  min_ele = float('inf')
  min_index = 0
  for i in range(len(arr)):
    if arr[i] < min_ele:
      min_ele = arr[i]
      min_index = i
  return min_index

print(f"Minimum element index in A  = {argmin(A)}")
print(f"Minimum element index using np.argmin in A = {np.argmin(A)}")
```
